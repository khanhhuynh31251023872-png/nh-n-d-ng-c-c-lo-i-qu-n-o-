```python
import tensorflow as tf
import numpy as np
import gradio as gr
from PIL import Image
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
class_names = [
    "Áo thun / T-shirt",
    "Quần dài / Trouser",
    "Áo len / Pullover",
    "Váy / Dress",
    "Áo khoác / Coat",
    "Dép / Sandal",
    "Áo sơ mi / Shirt",
    "Giày sneaker",
    "Túi xách / Bag",
    "Boot / Ankle boot"
]
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Chuẩn hóa ảnh về 0 - 1
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# One-hot label
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Train:", x_train.shape)
print("Test:", x_test.shape)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation="relu"),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.3),
    Dense(10, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    x_train,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_data=(x_test, y_test_cat)
)
```

```python
loss, acc = model.evaluate(x_test, y_test_cat)
print("Độ chính xác của model:", round(acc * 100, 2), "%")
```

```python
import cv2
import numpy as np
from PIL import Image

def preprocess_drawing(image):
    # Nếu Gradio trả về dict từ ImageEditor
    if isinstance(image, dict):
        if "composite" in image:
            image = image["composite"]
        elif "background" in image:
            image = image["background"]
        else:
            raise ValueError("Không lấy được ảnh từ canvas")

    # Chuyển sang ảnh xám
    if len(image.shape) == 3:
        img = Image.fromarray(image).convert("L")
    else:
        img = Image.fromarray(image).convert("L")

    img = np.array(img)

    # Đảo màu nếu nền trắng, nét đen
    # Sau bước này: nền đen, nét trắng
    if img.mean() > 127:
        img = 255 - img

    # Lọc nhiễu và làm rõ nét
    _, img = cv2.threshold(img, 30, 255, cv2.THRESH_BINARY)

    # Tìm vùng có nét vẽ
    coords = cv2.findNonZero(img)

    if coords is None:
        return None

    x, y, w, h = cv2.boundingRect(coords)

    # Cắt vùng hình vẽ
    cropped = img[y:y+h, x:x+w]

    # Tạo ảnh vuông
    size = max(w, h)
    square = np.zeros((size, size), dtype=np.uint8)

    x_offset = (size - w) // 2
    y_offset = (size - h) // 2

    square[y_offset:y_offset+h, x_offset:x_offset+w] = cropped

    # Thêm viền đen cho giống Fashion MNIST
    square = cv2.copyMakeBorder(
        square,
        20, 20, 20, 20,
        cv2.BORDER_CONSTANT,
        value=0
    )

    # Resize về 28x28
    img_28 = cv2.resize(square, (28, 28), interpolation=cv2.INTER_AREA)

    # Làm nét mềm hơn một chút
    img_28 = cv2.GaussianBlur(img_28, (3, 3), 0)

    # Chuẩn hóa
    img_28 = img_28 / 255.0

    return img_28


def predict_clothing(image):
    if image is None:
        return "Bạn chưa vẽ hình nào.", {}

    try:
        img_28 = preprocess_drawing(image)

        if img_28 is None:
            return "Bạn chưa vẽ hình nào.", {}

        input_img = img_28.reshape(1, 28, 28, 1)

        prediction = model.predict(input_img, verbose=0)[0]

        predicted_index = np.argmax(prediction)
        predicted_label = class_names[predicted_index]
        confidence = prediction[predicted_index] * 100

        # Sắp xếp xác suất từ cao xuống thấp
        sorted_indices = np.argsort(prediction)[::-1]

        top3_text = "TOP 3 DỰ ĐOÁN:\n\n"
        for rank, idx in enumerate(sorted_indices[:3], start=1):
            top3_text += f"{rank}. {class_names[idx]}: {prediction[idx] * 100:.2f}%\n"

        final_text = f"""
Kết quả dự đoán cao nhất: {predicted_label}

Độ tin cậy: {confidence:.2f}%

{top3_text}
"""

        result = {
            class_names[i]: float(prediction[i])
            for i in range(10)
        }

        return final_text, result

    except Exception as e:
        return f"Lỗi xử lý ảnh: {str(e)}", {}
```
